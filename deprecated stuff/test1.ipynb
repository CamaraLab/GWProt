```python
import os
os.environ['OPENBLAS_NUM_THREADS'] = '1' 
import multiprocessing
import numpy as np

def helper(x):
    np.random.seed(x)
    A = np.random.rand(1000,1000)
    return np.mean(A@A@A)
```

```python
%%time
with multiprocessing.Pool(16) as pool:
    results = pool.imap(helper, range(200))
    for r in results:
        print(r)
```

```python
from multiprocessing import Pool, Manager

class LargeObject:
    def __init__(self, data):
        self.data = data  # Large list or data to be shared

    @staticmethod
    def _worker(y):
        shared_data, x = y
        # Static method used by each worker process
        return sum([d * x for d in shared_data])

    def parallel_compute(self, values, num_processes=4):
        # Create the Manager and share the object's data
        with Manager() as manager:
            shared_data = manager.list(self.data)  # Create manager-backed shared data

            # Perform the parallel computation using multiprocessing Pool
            with Pool(processes=num_processes) as pool:
                # Pass the shared data and value to each worker
                results = pool.imap(LargeObject._worker, [(shared_data, v) for v in values])

        return results
```

```python
# Create a normal large object
large_object = LargeObject([1, 2, 3, 4, 5] * 1000000)

# Perform parallel computations using object method (Manager created internally)
results = large_object.parallel_compute(range(10), num_processes=4)
print(list(results))
```
