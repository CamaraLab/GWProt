```python
import ot
import numpy as np
import math
import scipy
```

```python
def stress(A,a,B,b,T):

    stress1 = np.einsum('ik,il->i',T,(np.einsum('ij,ij->ij',A,A) @T) )   + T @ np.einsum('kl,kl->kl',B,B) @b  -(2 * np.einsum('ab,ab->a', A @T @B, T))
    stress2 = np.einsum('kj,lj->j' ,T @ np.einsum('kl,kl->kl',B,B), T) +a.T @ np.einsum('ij,ij->ij',A,A) @T -(2 * np.einsum('ab,ab->b', A @T @B, T))

    return stress1, stress2
```

```python
X = np.random.rand(50,3)
Y = np.random.rand(50,3)
A = scipy.spatial.distance.squareform(scipy.spatial.distance.pdist(X))
B = scipy.spatial.distance.squareform(scipy.spatial.distance.pdist(Y))

a = np.random.rand(50)
a = a / np.sum(a)
b = np.random.rand(50)
b = b / np.sum(b)


T, log = ot.gromov_wasserstein(C1  =A, C2 = B, p=a, q=b, log = True)
d = log['gw_dist']
```

```python
s1, s2 = stress(A,a,B,b,T)
np.sum(s1), np.sum(s2)
```
